```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

r=requests.get("https://www.webscraper.io/test-sites/e-commerce/allinone/computers/laptops")

c=r.content
```

```python


soup=BeautifulSoup(c,"html.parser")

print(soup.prettify())


all=soup.find_all("div",{"class":"col-sm-4 col-lg-4 col-md-4"})
```

```python
l=[]

for item in all:
    d={}
    d["Laptop Name"]=item.find("a").text[:-3]
    arr=item.find("p").text.split(',')
    d["price"]=item.find("h4",{"class":"pull-right price"}).text
    try:
        screen=arr[1]
        d["Screen size"]=screen[1:]
    except:
        d["Screen size"]=None
    try:
        processor=arr[2]
        d["Processor type"]=processor[1:]
    except:
        d["Processor type"]=None
    try:
        ram=arr[3]
        d["RAM Size"]=ram[1:]
    except:
        d["RAM Size"]=None
    try:
        drive_space=arr[4]
        d["Driver Space"]=drive_space[1:]
    except:
        d["Driver Space"]=None
    try:
        os_type=arr[5]
        d["OS Type"]=os_type[1:]
    except:
        d["OS Type"]=None
    l.append(d)

```

```python
df=pd.DataFrame(l)
df
df.to_csv('Output.csv')

```
